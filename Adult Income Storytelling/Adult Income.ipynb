## EDA

```python
df_age = df.set_index('age')
```

```python
df_age['educational-num'].value_counts(ascending=False)
```

This is very interesting. A large number of people in this data set's highest education is either some college or they are a high school graduate. College is not for everyone, but this is interesting to see off the bat.

```python
df_age['hours-per-week'].value_counts(ascending=False)
```

American society runs on having employees work 40 hours per week, so it is no surprise that 40 hours is the highest value count by far. Interesting to see the lower few values...who is working 79-94 hours a week?

```python
df['race'].value_counts()
```

```python
df.education.value_counts().plot(kind = "pie")
```

Some college, bachelors and high school grad are the largest parts of the pie.

```python
corr = df._get_numeric_data().corr()
sns.heatmap(
    corr, 
    xticklabels=corr.columns.values,
    yticklabels=corr.columns.values
);
```

Numerical Values have very low correlation.

```python
sns.countplot(x='income', data=df)
```

#### Income Based on Marital Status

```python
sns.countplot(x='income', hue='marital-status', data=df)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

#### Income Based on Occupation

```python
sns.countplot(x='income', hue='occupation', data=df)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

#### Income Based on Gender

```python
sns.countplot(x='income', hue='gender', data=df)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

#### Hours Ages are Working

```python
sns.scatterplot(x=df['hours-per-week'], y=df['age'], hue=df['income'])
plt.figure(figsize=(12,9))
```
